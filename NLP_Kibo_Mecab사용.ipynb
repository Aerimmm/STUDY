```python
from konlpy.tag import Mecab
```

```python
stop_words = " ( ) , / ? 세 가 이 은 한 에 하 고 을 를 인 듯 과 와 네 들 및 듯 지 임 계 화 적 식 브 변 의 어 온 류 용 및 형 급 는 것 등 케 " # 불용어 정의
stop_words =stop_words.split(' ') # 띄어쓰기대로 분리
stop_words
```

```python
# 형태소 분석기 Mecab 를 이용한 토큰화 작업

mecab = Mecab()
    
df['tokenized'] = df['기술명'].apply(mecab.morphs)
tokenized_data = df['tokenized'].apply(lambda x: [item for item in x if item not in stop_words])
tokenized_data
```

# FastText 을 이용한 Word Embedding

## Word Embedding
- 단어를 Embedding 하는 것. <br>
단어 Embedding 이란 단어의 의미를 고려하여 좀 더 조밀한 차원의 단어를 벡터로 표현하는 것이다. <br>
LSA, Word2Vec, GloVe, FastText 등의 단어임베딩 모형이 있다.

## FastText
- 페이스북에서 개발
- Word2Vec은 단어를 쪼개질 수 없는 단위로 생각하지만,<br>
 FastText는 하나의 단어 안에도 여러 단어들이 존재한다고 간주하고 내부단어 즉, 서브워드(Subword) 를 고려하여 학습한다.

```python
### FastText 학습 
from gensim.models import FastText
ft_model = FastText(tokenized_data, size=100, window=5, min_count=5, workers=10, sg=1) # 모델에 토큰을 넣는다. 근처 5개를 보도록 함, 5번 이하로 등장하는 단어는 X, 4번의 프로세스, Skip Gram 사용
```

### 하이퍼 파라미터 구성
- Size = 워드 벡터의 특징 값. Embedding 된 단어의 차원
- Window = Context 윈도우 크기. 주변 단어를 몇 개 볼 건지
- min_count = 단어 최소 빈도 수 제한
- workers = 학습을 위한 프로세스 수
- sg = 0은 CBOW, 1은 Skip-Gram <br>
CBOW 는 주변 단어들을 통해 중간의 단어를 예측하는 모델 <br>
Skip-Gram 은 중심 단어를 통해 주변 단어를 예측하는 모델


```python
ft_model.wv.vectors.shape # 총 14204개의 단어가 존재하며 100차원으로 구성
```
