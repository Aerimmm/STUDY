# Matplotlib 공부

### Data Dictionary
- People  
- 1. ID: Customer's unique identifier (고객식별id)
- 2. Year_Birth: Customer's birth year (출생년도)
- 3. Education: Customer's education level (학력)
- 4. Marital_Status: Customer's marital status (결혼상태)
- 5. Income: Customer's yearly household income (수입)
- 6. Kidhome: Number of children in customer's household (자녀 수)
- 7. Teenhome: Number of teenagers in customer's household (10대 자녀 수)
- 8. Dt_Customer: Date of customer's enrollment with the company (회원가입날짜)
- 9. Recency: Number of days since customer's last purchase (마지막구매 후 경과일자)
- 10. Complain: 1 if the customer complained in the last 2 years, 0 otherwise (지난 2년간 


- Products

- 1. MntWines: Amount spent on wine in last 2 years (지난 2년간 와인 구매량)
- 2. MntFruits: Amount spent on fruits in last 2 years (지난 2년간 과일 구매량)
- 3. MntMeatProducts: Amount spent on meat in last 2 years (지난 2년간 육류 구매량)
- 4. MntFishProducts: Amount spent on fish in last 2 years (지난 2년간 수산물 구매량)
- 5. MntSweetProducts: Amount spent on sweets in last 2 years (지난 2년간 간식류 구매량)
- 6. MntGoldProds: Amount spent on gold in last 2 years (지난 2년간 귀금속 구매량)

## Histogram

```python
# 기본 틀 만들기
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)

# 히스토그램 그래프 만들기
axes1.hist(df['Kidhome'],bins = 6) # 가정의 자녀 수 히스토그램
# bins 는 x축 칸을 조절해준다.
axes1.set_title("고객 가정의 자녀 수 히스토그램") # 제목설정
axes1.set_xlabel("자녀수")
axes1.set_ylabel("빈도")
plt.show()
```

## Box Plot

```python
# box 그래프그리기 (이산형 변수와 연속형 변수를 함께 사용한다)
# ㅇㅣ산형 : 학력
# 연속형 : 수입
# df Null값 0 으로 대체한다.
df['Income'] = df['Income'].fillna(0)

boxplot = plt.figure()
axes3 = boxplot.add_subplot(1,1,1)
axes3.boxplot([df[df['Education'] == 'Basic']['Income'],
               df[df['Education'] == 'Graduation']['Income'],
               df[df['Education'] == '2n Cycle']['Income'],
               df[df['Education'] == 'Master']['Income'],
               df[df['Education'] == 'PhD']['Income']],
               labels = ['고졸','대졸','2nCycle','석사','박사']
              )

axes3.set_xlabel('학력')
axes3.set_ylabel('수입')
axes3.set_title('학력에 따른 수입 박스그래프')
plt.show()
```

```python
# df피벗
# 교육, 수입부분만 뽑기
df_pv = df.iloc[:,0:5]
df_pv = df_pv.drop(['Year_Birth','Marital_Status'],axis = 1) #열 삭제시 axis=1
print(df_pv.head())
```

```python
# 교육상태를 칼럼으로 피벗
df_pv = df_pv.pivot_table(
    index = 'ID',
    columns = 'Education',
    values = 'Income')
print(df_pv.head())
```

```python
# df로 박스그래프 만들기
fig3, ax3 = plt.subplots()
ax3 = df_pv.plot.box(ax=ax3)
plt.show()
```

## PiChart

```python
# 학력 카운트
count = df.groupby('Education').Education.count()

# 전체 고객 중 해당 학력을 가지고 있는 비율 구하기
ratio = count/sum(count)

# 그래프에 표시할 인덱스
labels = count.index

#파이그래프 그리기
pie = plt.pie(ratio, labels=labels, autopct = '%.1f%%')
plt.show()
```

```python
# explode = 파이를 띄울 간격
# shadow = True 파이 그림자 설정
# colors = 그래프 색 변경
pie = plt.pie(ratio, labels=labels, autopct = '%.1f%%', explode = [0,0,0.05,0,0.1], shadow = True,
             colors = ['#FAF4B7', '#ECC5FB', '#C6EBC5','#CDF0EA','#FFFAE7'])
plt.show()
```

# 다변량 그래프 그리기

```python
# 다변량 그래프 (3개 이상의 변수 사용)

#학력 부분 숫자로 바꾸기
def recode_edu (education_level):
    if education_level == 'Basic':
        return 0
    if education_level == 'Graduation':
        return 1
    if education_level == '2nCycle':
        return 2
    if education_level == 'Master':
        return 3
    else :
        return 4
    
df['Edu_level'] = df['Education'].apply(recode_edu)
df.head()
```

- x축: MntMeatProducts(육류구매량)
- y축: MntWines(와인구매량)
- c(컬러): Edu_level(학력)
- s(사이즈): NumWebPurchases(웹사이트구매횟수)

```python
# 그래프틀 생성
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)

# 산점도 다변량 그래프 그리기
axes1.scatter(
    x = df['MntMeatProducts'],
    y = df['MntWines'],
    c = df['Edu_level'],
    s = df['NumWebPurchases']*10,
    alpha = 0.5) #alpha 는 투명도 조절

axes1.set_title("다변량 그래프 (학력별 다른 색상, 웹사이트 구매횟수는 사이즈별)")
axes1.set_xlabel("육류구매량")
axes1.set_ylabel("와인구매량")
plt.show()
```
